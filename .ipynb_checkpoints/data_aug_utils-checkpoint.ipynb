```python
# resize
# crop
# flip
# brightness & contrast

import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
from matplotlib.pyplot import imshow
```

```python
name = './gugong.jpg'
img_string = tf.read_file(name)
img_decoded = tf.image.decode_image(img_string)
sess = tf.Session()
img_decoded_val = sess.run(img_decoded)
print(img_decoded_val.shape)
%matplotlib inline
imshow(img_decoded_val)
```

```python
# resize

# tf.image.resize_area
# tf.image.resize_bicubic
# tf.iamge.resize_nearest_neighbor

name = './gugong.jpg'
img_string = tf.read_file(name)
img_decoded = tf.image.decode_image(img_string)
img_decoded = tf.reshape(img_decoded, [1, 675, 1200, 3])

resize_img = tf.image.resize_bicubic(img_decoded, [1350, 2400])

sess = tf.Session()
img_decoded_val = sess.run(resize_img)
img_decoded_val = img_decoded_val.reshape((1350, 2400, 3))
img_decoded_val = np.asarray(img_decoded_val, np.uint8)

print(img_decoded_val.shape)
%matplotlib inline
imshow(img_decoded_val)
```

```python
# crop
# tf.image.pad_to_bounding_box
# tf.image.crop_to_bounding_box
# tf.random_crop

name = './gugong.jpg'
img_string = tf.read_file(name)
img_decoded = tf.image.decode_image(img_string)
img_decoded = tf.reshape(img_decoded, [1, 675, 1200, 3])

padded_img = tf.image.pad_to_bounding_box(img_decoded, 50, 100, 500, 800)

sess = tf.Session()
img_decoded_val = sess.run(padded_img)
img_decoded_val = img_decoded_val.reshape((700, 800, 3))
img_decoded_val = np.asarray(img_decoded_val, np.uint8)

print(img_decoded_val.shape)
%matplotlib inline
imshow(img_decoded_val)
```
